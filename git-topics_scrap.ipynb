```python
for topic_url  in df['Topic_urls'] :
    html_doc=requests.get(topic_url).text
    soup=BeautifulSoup(html_doc,"html.parser")
    Username=[]
    Repo_name=[]
    Stars=[]
    repo_urls=[]
    base_url="https://github.com"
    arti_para="border rounded color-shadow-small color-bg-subtle my-4"
    articles=soup.find_all("article",{"class":arti_para})

    for article in articles:
        anchor_tags=article.find_all('h3',{'class':'f3 color-fg-muted text-normal lh-condensed'})
        for anchor in anchor_tags:
            Username.append(anchor.find('a',{'class':'Link'}).text.strip())
            Repo_name.append(anchor.find('a',{'class':'Link text-bold wb-break-word'}).text.strip())
            repo_urls.append(base_url+anchor.find('a',{'class':'Link text-bold wb-break-word'})['href'])
        stars_count=article.find_all('span',{'class':"Counter js-social-count"})   
        for counts in stars_count:
            count=counts.text
            if count[-1]=="k":
                Stars.append(int(float(count[:-1])*1000)) 
            else:
                Stars.append(count)

    df1=pd.DataFrame({"Username":Username,"Repo_name":Repo_name,"Stars":Stars,"Repo_urls":repo_urls},index=None)
    for i in range(len(df['Topic_title_tag'])):
        df1.to_csv("{}.csv".format(df['Topic_title_tag'][i]))
```
